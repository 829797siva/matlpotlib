```python
#import the libraires
import matplotlib.pyplot as plt
import numpy as np
fig=plt.figure()
ax=fig.add_subplot(111,projection='3d')
x=np.random.rand(5)
y=np.random.rand(5)
z=np.random.rand(5)
ax.scatter(x,y,z)
plt.show()
```
