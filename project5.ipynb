5. Histogram of Age Distribution

Goal: Plot a histogram for ages of people attending an event.

Extra: Customize bins and add a density curve.

Skills: Histogram, data distribution understanding.

```python
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns  # For smooth density curve

# Step 1: Generate sample age data
np.random.seed(42)  # For reproducibility
ages = np.random.normal(loc=30, scale=10, size=200)  # mean=30, std=10
ages = np.clip(ages, 10, 60)  # Restrict ages between 10 and 60

# Step 2: Plot histogram
plt.figure(figsize=(8,5))
plt.hist(ages, bins=10, color='skyblue', edgecolor='black', alpha=0.7, density=True)

# Step 3: Add a density (KDE) curve
sns.kdeplot(ages, color='red', linewidth=2, label='Density Curve')

# Step 4: Add titles and labels
plt.title('Age Distribution of Event Attendees', fontsize=14)
plt.xlabel('Age', fontsize=12)
plt.ylabel('Density', fontsize=12)
plt.grid(axis='y', linestyle='--', alpha=0.6)
plt.legend()

# Step 5: Display the histogram
plt.show()
```
