```python
#import the libraires
import matplotlib.pyplot as plt
labels=["python","java","c++","data"]
sizes=[40,50,80,20]
plt.pie(sizes,labels=labels,autopct="%1.1f%%")
plt.title("pie chart")
plt.show()
```
