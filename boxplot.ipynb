```python
#import the libraires
import matplotlib.pyplot as plt
import numpy as np
data=np.random.rand(100)
plt.boxplot(data)
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.title('Line Plot')
plt.show()
```
