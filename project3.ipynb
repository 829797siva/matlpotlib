3. Movie Ratings Visualization

Goal: Compare ratings of 5 movies from different platforms (IMDB, Rotten Tomatoes, etc.).

Tasks:

Use grouped bar charts.

Add legend and x-ticks with movie names.

Skills: Grouped bar chart, tick rotation.

```python
import numpy as np
import matplotlib.pyplot as plt

# Step 1: Movie names and ratings
movies = ['Inception', 'Interstellar', 'Avatar', 'Titanic', 'Joker']
imdb = [8.8, 8.6, 7.8, 7.9, 8.5]
rotten_tomatoes = [87, 72, 82, 88, 69]
metacritic = [74, 74, 83, 75, 59]

# Step 2: Set position of bars on X-axis
bar_width = 0.25
x = np.arange(len(movies))  # Base positions for movies

# Step 3: Plot grouped bars
plt.figure(figsize=(9,6))
plt.bar(x - bar_width, imdb, width=bar_width, label='IMDB', color='skyblue', edgecolor='black')
plt.bar(x, rotten_tomatoes, width=bar_width, label='Rotten Tomatoes', color='salmon', edgecolor='black')
plt.bar(x + bar_width, metacritic, width=bar_width, label='Metacritic', color='lightgreen', edgecolor='black')

# Step 4: Add titles and labels
plt.title(' Movie Ratings Comparison Across Platforms', fontsize=14)
plt.xlabel('Movies', fontsize=12)
plt.ylabel('Ratings', fontsize=12)

# Step 5: Customize x-ticks and legend
plt.xticks(x, movies, rotation=30)
plt.legend()
plt.grid(axis='y', linestyle='--', alpha=0.6)

# Step 6: Display chart
plt.tight_layout()
plt.show()
```
