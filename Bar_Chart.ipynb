```python
import matplotlib.pyplot as plt

subject=["math","java","c","english"]
marks=[77,88,99,45]

plt.bar(subject,marks)

plt.title("subject vs marks")
plt.xlabel("subject")
plt.ylabel("marks")
plt.show
```
