2. Sales Data Analysis

Goal: Plot monthly sales data using line and bar charts.

Tasks:

Create months = ['Jan', 'Feb', 'Mar', ...]

Plot line chart for sales.

Add title, labels, and grid.

Skills: Line plot, axis labeling, gridlines.

```python
# Import matplotlib for plotting
import matplotlib.pyplot as plt

# Step 1: Create sample monthly sales data
months = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun',
          'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']

sales = [12000, 15000, 17000, 13000, 16000, 18000,
         20000, 19500, 21000, 22000, 25000, 27000]

# Step 2: Plot Line Chart
plt.figure(figsize=(8,5))
plt.plot(months, sales, color='blue', marker='o', linestyle='-', linewidth=2)
plt.title('Monthly Sales Data - Line Chart')
plt.xlabel('Month')
plt.ylabel('Sales Amount ($)')
plt.grid(True, linestyle='--', alpha=0.6)
plt.show()

# Step 3: Plot Bar Chart
plt.figure(figsize=(8,5))
plt.bar(months, sales, color='skyblue', edgecolor='black')
plt.title('Monthly Sales Data - Bar Chart')
plt.xlabel('Month')
plt.ylabel('Sales Amount ($)')
plt.grid(axis='y', linestyle='--', alpha=0.6)
plt.show()
```
