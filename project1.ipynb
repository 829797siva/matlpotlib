1. Student Marks Visualization

Goal: Read a CSV file containing student names and marks in 3 subjects, and plot:

Bar chart for total marks.

Pie chart for subject-wise average.

Skills: Bar plot, pie chart, labels, legends.

```python
# Import required libraries
import pandas as pd
import matplotlib.pyplot as plt

# Step 1: Create a sample CSV file (you can skip this if you already have one)
data = {
    'Name': ['Alice', 'Bob', 'Charlie', 'David', 'Eva'],
    'Math': [85, 78, 92, 70, 88],
    'Science': [90, 82, 85, 75, 95],
    'English': [80, 75, 89, 68, 92]
}

df = pd.DataFrame(data)
df.to_csv('student_marks.csv', index=False)  # Save CSV file

# Step 2: Read the CSV file
df = pd.read_csv('student_marks.csv')
print("ðŸ“„ Student Data:")
print(df)

# Step 3: Calculate total marks for each student
df['Total'] = df[['Math', 'Science', 'English']].sum(axis=1)

# Step 4: Plot bar chart for total marks
plt.figure(figsize=(8,5))
plt.bar(df['Name'], df['Total'], color='skyblue', edgecolor='black')
plt.title('Total Marks of Each Student')
plt.xlabel('Student Name')
plt.ylabel('Total Marks')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()

# Step 5: Calculate average marks per subject
subject_avg = df[['Math', 'Science', 'English']].mean()

# Step 6: Plot pie chart for subject-wise average
plt.figure(figsize=(6,6))
plt.pie(subject_avg, labels=subject_avg.index, autopct='%1.1f%%', startangle=90, colors=['#ff9999','#66b3ff','#99ff99'])
plt.title('Average Marks by Subject')
plt.show()
```
