```python
import matplotlib.pyplot as plt
import numpy as np
from mpl_toolkits.mplot3d import Axes3D
```

```python

# 1️⃣ LINE PLOT
x = [1, 2, 3, 4, 5]
y = [2, 4, 6, 8, 10]
plt.figure()
plt.plot(x, y, color='blue', marker='o', linestyle='--')
plt.title("1. Line Plot")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.show()
```

```python
# 2️⃣ BAR CHART
subjects = ['Math', 'Science', 'English', 'History']
marks = [85, 90, 75, 95]
plt.figure()
plt.bar(subjects, marks, color='green')
plt.title("2. Bar Chart")
plt.xlabel("Subjects")
plt.ylabel("Marks")
plt.show()
```

```python
# 3️⃣ HORIZONTAL BAR CHART
plt.figure()
plt.barh(subjects, marks, color='orange')
plt.title("3. Horizontal Bar Chart")
plt.xlabel("Marks")
plt.ylabel("Subjects")
plt.show()
```

```python
# 4️⃣ PIE CHART
languages = ['Python', 'Java', 'C++', 'JavaScript']
usage = [40, 25, 20, 15]
plt.figure()
plt.pie(usage, labels=languages, autopct='%1.1f%%', startangle=140)
plt.title("4. Pie Chart")
plt.show()
```

```python
# 5️⃣ SCATTER PLOT
height = [150, 160, 170, 180, 190]
weight = [50, 55, 65, 75, 85]
plt.figure()
plt.scatter(height, weight, color='red', marker='^')
plt.title("5. Scatter Plot")
plt.xlabel("Height (cm)")
plt.ylabel("Weight (kg)")
plt.show()
```

```python
# 6️⃣ HISTOGRAM
data = np.random.randn(1000)
plt.figure()
plt.hist(data, bins=20, color='purple', edgecolor='black')
plt.title("6. Histogram")
plt.xlabel("Value Range")
plt.ylabel("Frequency")
plt.show()
```

```python
# 7️⃣ BOX PLOT
data = [20, 22, 23, 25, 29, 30, 32, 35, 36, 45, 100]
plt.figure()
plt.boxplot(data)
plt.title("7. Box Plot")
plt.ylabel("Values")
plt.show()
```

```python
# 8️⃣ AREA (STACK) PLOT
days = [1, 2, 3, 4, 5]
sleep = [7, 8, 6, 11, 7]
study = [2, 3, 4, 3, 2]
work = [8, 7, 8, 6, 8]
plt.figure()
plt.stackplot(days, sleep, study, work, labels=['Sleep', 'Study', 'Work'])
plt.legend(loc='upper left')
plt.title("8. Stack (Area) Plot")
plt.xlabel("Days")
plt.ylabel("Hours")
plt.show()
```

```python
# 9️⃣ STEM PLOT
x = [1, 2, 3, 4, 5]
y = [1, 4, 9, 16, 25]
plt.figure()
plt.stem(x, y)
plt.title("9. Stem Plot")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.show()
```

```python
# 🔟 ERROR BAR PLOT
x = [1, 2, 3, 4, 5]
y = [10, 20, 25, 30, 40]
error = [1, 2, 1.5, 3, 2.5]
plt.figure()
plt.errorbar(x, y, yerr=error, fmt='o', color='magenta', ecolor='black', capsize=5)
plt.title("10. Error Bar Plot")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.show()
```

```python
# 11️⃣ VIOLIN PLOT
data = [np.random.normal(0, std, 100) for std in range(1, 4)]
plt.figure()
plt.violinplot(data, showmeans=True)
plt.title("11. Violin Plot")
plt.show()
```

```python
# 12️⃣ HEATMAP
data = np.random.rand(5, 5)
plt.figure()
plt.imshow(data, cmap='hot', interpolation='nearest')
plt.title("12. Heatmap")
plt.colorbar()
plt.show()
```

```python
# 13️⃣ MULTIPLE SUBPLOTS
x = [1, 2, 3, 4, 5]
y1 = [2, 4, 6, 8, 10]
y2 = [1, 3, 5, 7, 9]
plt.figure()
plt.subplot(1, 2, 1)
plt.plot(x, y1, color='blue')
plt.title("13A. Subplot 1")
plt.subplot(1, 2, 2)
plt.plot(x, y2, color='red')
plt.title("13B. Subplot 2")
plt.show()
```

```python
# 14️⃣ 3D SCATTER PLOT
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
x = np.random.rand(50)
y = np.random.rand(50)
z = np.random.rand(50)
ax.scatter(x, y, z, color='green')
ax.set_title("14. 3D Scatter Plot")
plt.show()
```
