4. Boxplot of Exam Scores

Goal: Show the distribution of scores across multiple subjects.

Skills: Boxplots, identifying outliers.

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

# Step 1: Create sample data
np.random.seed(42)  # For reproducibility

data = {
    'Math': np.random.randint(50, 100, 25),
    'Science': np.random.randint(40, 95, 25),
    'English': np.random.randint(45, 98, 25),
    'History': np.random.randint(55, 100, 25),
    'Art': np.random.randint(60, 100, 25)
}

df = pd.DataFrame(data)
print("ðŸ“Š Sample Exam Scores Data:")
print(df.head())

# Step 2: Plot boxplot
plt.figure(figsize=(8,6))
plt.boxplot(df.values, labels=df.columns, patch_artist=True,
            boxprops=dict(facecolor='lightblue', color='black'),
            medianprops=dict(color='red', linewidth=2))

# Step 3: Add title and labels
plt.title('Distribution of Exam Scores Across Subjects', fontsize=14)
plt.xlabel('Subjects', fontsize=12)
plt.ylabel('Scores', fontsize=12)
plt.grid(axis='y', linestyle='--', alpha=0.6)

# Step 4: Display the boxplot
plt.show()
```
